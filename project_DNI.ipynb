```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df_hr['Direct Normal [W/m^2]'], model='additive', period=24)
trend = result.trend.dropna()
seasonal = result.seasonal.dropna()
residual = result.resid.dropna()

# Plot the decomposed components
plt.figure(figsize=(10,6))



plt.subplot(3, 1, 1)
plt.plot(trend, label='Trend')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(seasonal, label='Seasonal')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(residual, label='Residuals')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
X = DNI_series.values
diff = list()
hours_in_days = 24
for i in range(hours_in_days, len(X)):
 value = X[i] - X[i - hours_in_days]
 diff.append(value)
plt.figure(figsize=(25,8))
plt.plot(diff[-1000:])
plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARMA
```

```python
num_lags = 24
acf_vals = acf(diff)
plt.bar(range(num_lags), acf_vals[:num_lags])
```

```python
plot_acf(diff, lags=24)
plt.show()
```

```python
plot_pacf(diff, lags=24)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(DNI_series.values, order=(2,0,3))
model_fit = model.fit()

# Print a summary of the model fit
print(model_fit.summary())
```

```python
start_index = len(DNI_series.values) - 100
end_index = len(DNI_series.values) - 1
predictions = model_fit.predict(start=start_index, end=end_index)

# Plot the predictions
plt.figure(figsize=(25, 8))
plt.plot(DNI_series.values[start_index:], label='Actual')
plt.plot(predictions, color='red', label='Predicted')
plt.xlabel("Index", fontsize=18)
plt.ylabel("(DNI)", fontsize=18)
plt.legend()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
from datetime import datetime
from datetime import timedelta
```

```python
DNI_series=df_hr['Direct Normal [W/m^2]']
train_end=datetime(2013,12,24,23,0)
test_end= datetime(2013,12,31,23,0)
train_data=DNI_series[:train_end]
test_data=DNI_series[train_end+timedelta(hours=1):test_end]
test_data
```

```python
holt_winter = ExponentialSmoothing(train_data.values,
                                      seasonal='add',
                                      seasonal_periods=24).fit(smoothing_level=0,
                                                                  smoothing_trend=0,
                                                                  smoothing_seasonal=.5,
                                                                  optimized=False)
```

```python
plt.figure(figsize=(25,8))
start_date=datetime(2013,12,21)
end_date=datetime(2013,12,31,17,0)
lim_DNI_series= DNI_series[start_date:end_date]

predictions=holt_winter.forecast(len(test_data))
predictions=pd.Series(predictions, index=test_data.index)

plt.plot(lim_DNI_series)
plt.plot(predictions,'r')


plt.xlabel("Date", fontsize=18)
plt.ylabel("W/m^2", fontsize=18)
plt.title('Forcast DNI using Exponential Smoothing',fontsize=20)

plt.legend(('Date','predictions'), fontsize=16)
plt.show()
```

```python
residuals=test_data-predictions
print("Root Mean Squared Error:", np.sqrt(np.mean(residuals**2)))
```
