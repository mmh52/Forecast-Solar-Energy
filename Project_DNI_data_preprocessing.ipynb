```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('LDNI.csv')
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5))
plt.plot(df['Direct Normal [W/m^2]'])
```

```python
df.info()
```

```python
df=df.replace(-99999, np.NaN)
df.fillna(method='ffill')
```

```python
plt.figure(figsize=(25,5))
plt.plot(df['Direct Normal [W/m^2]'])
```

```python
df['Direct Normal [W/m^2]'] = (df['Direct Normal [W/m^2]']==np.NaN).where(df['Direct Normal [W/m^2]']>1100,df['Direct Normal [W/m^2]'])
df
```

```python
df.info()
```

```python
df['Direct Normal [W/m^2]'] = pd.to_numeric(df['Direct Normal [W/m^2]'],errors = 'coerce')
```

```python
plt.figure(figsize=(25,5))
plt.plot(df['Direct Normal [W/m^2]'])
```

```python
df_hr=df.resample('60T').mean().interpolate('time')
df_hr
```

```python
df_hr=df_hr.reset_index()
df_hr.to_csv('DNI_hr.csv', index=False)
```
