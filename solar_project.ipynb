```python
df_hr=df_hr.reset_index()
df_hr.set_index('Date', inplace=True)
DNI_series = df_hr['Direct Normal [W/m^2]']
DNI_series

GH_series = df_hr['Global Horizontal [W/m^2]']
GH_series




# train_data = DNI_series[:train_end]
# test_data = DNI_series[train_end + timedelta(days=1):test_end]

train_features=GH_series[:train_end]
test_features=GH_series[train_end + timedelta(hours=1):test_end]



from statsmodels.tsa.statespace.sarimax import SARIMAX
model_s = SARIMAX(train_data, exog = train_features, order=(2, 0, 3), seasonal_order=(2, 0, 0, 24))
model_fit_s = model_s.fit()

# Print a summary of the model fit
print(model_fit_s.summary())
```

```python
#get the predictions and residuals
predictions = model_fit_s.forecast(len(test_data),exog = test_features)
predictions = pd.Series(predictions, index=test_data.index)
residuals = test_data - predictions

plt.figure(figsize=(25,8))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)


#plot prediction
start_date = datetime(2013,12,21)
end_date = datetime(2013,12,31,23,0)
lim_DNI_series = DNI_series[start_date:end_date]

plt.figure(figsize=(25,8))

plt.plot(lim_DNI_series)
plt.plot(predictions,'r')

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('DNI', fontsize=20)
plt.ylabel('W/m^2', fontsize=16)
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```

```python
GH_series = df_hr['Global Horizontal [W/m^2]']

rolling_predictions = test_data.copy()
for train_end in test_data.index:
    train_features= GH_series[:train_end-timedelta(hours=1)]
    train_data = DNI_series[:train_end-timedelta(hours=1)]
    
    model = SARIMAX(train_data,exog = train_features, order=(2, 0, 3), seasonal_order=(2, 0, 0, 24))
    model_fit = model.fit()

    pred = model_fit.forecast(steps=1,exog=GH_series[train_end:train_end])
    rolling_predictions[train_end] = pred
```

```python
rolling_residuals = test_data - rolling_predictions
plt.figure(figsize=(25,8))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Forecast Residuals Using SARIMAX Model', fontsize=20)
plt.ylabel('Error', fontsize=16)



plt.figure(figsize=(25,8))

plt.plot(lim_DNI_series)
plt.plot(rolling_predictions,'r')

plt.legend(('Data', 'Predictions'), fontsize=20)

plt.title('Rolling Forecast of DNI Using SARIMAX Model', fontsize=20)
plt.ylabel('W/m^2', fontsize=16)
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(rolling_residuals**2)))
```
